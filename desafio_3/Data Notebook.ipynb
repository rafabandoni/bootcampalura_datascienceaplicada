#Conclusão

Através da realização desse estudo, pude concluir que:
- O consumo precoce de álcool é uma realidade em nosso país, provavelmente por fatores culturais - principalmente nos estados da região Sul;
- A companhia é um dos maiores indicadores para consumo precoce de álcool. Ter amigos que bebem aumenta drasticamente as chances de um menor de idade ingerir álcool.
- Alguns dados, como o trabalho antes dos 18 anos e morar com apenas um dos pais podem influenciar, mas bem menos.
- O consumo de cigarro também é um problema sério entre os menores de idade que bebem, 1 em cada 3 dos que consumiram álcool antes dos 18 anos também já experimentou cigarro.
- Quando estamos falando de outras drogas, apenas 15% dos que já consumiram álcool de maneira precoce já usaram outras drogas. Ao passo que esse número sobe para 37% se comparado com os que já fumaram cigarro - o cigarro precoce (que também é uma consequência do álcool precoce) parece ser a maior "porta de entrada" para outras drogas).

#Resumo

##Metodologia
A metodologia desse estudo consistiu em três etapas principais, utlizando a Amostra 1 da base PENSE do IBGE. Sendo:
- Entendimento do perfil dos respondentes que usavam álcool antes dos 18 anos.
- Cruzamento entre essa informação e índices financeiros e comportamentais.
- Entendimento da influência do uso precoce de álcool e o acesso a cigarro e outras drogas.

Abaixo, alguns resultados:

##Perfil

Em 15 dos 27 estados brasileiros, o consumo precoce de álcool é maioria entre os respondentes da pesquisa, o que mostra um problema a ser tratado pelos órgãos competentes em nosso país - haja visto que [o consumo precoce de álcool pode trazer consequências para o desenvolvimento](https://fcmsantacasasp.edu.br/os-riscos-do-consumo-de-alcool-na-adolescencia/).  
Para auxiliar na solução desse problema, começo traçando um possível perfil dos que consomem álcool antes dos 18 anos (os resultados e cruzamentos podem ser vistos na "Parte 1" do índice).  

**O que não importa?**  
Alguns fatores, por mais que possam parecer, não parecem ter tanta importância a primeira vista. São eles: O turno escolar, a cor ou raça declarada. Esses fatores mostraram pouca ou nenhuma correlação com o consumo precoce de álcool.

**O que importa?**  
Aparentemente, o estado importa - principalmente os estados do sul. Provavelmente por fatores culturais, é possível ver uma forte tendência do consumo precoce de álcool nos estados da região sul do nosso país.  
O sexo declarado tem um certo peso, mas muito pouco. Entre os que se declararam sexo feminino, 52% já ingeriu álcool antes dos 18 anos, contra 50% do sexo masculino.  
O trabalho antes dos 18 anos também tem certa influência. Entre os que trabalham, o consumo precoce de álcool é de 66%, contra 49% dos que não trabalham.

##Comportamento

Já os fatores comportamentais ou situacionais, como mostrados na parte 2 do índice, se mostraram de grande importância nesse aspecto. Da mesma forma, alguns não têm tanta influência, vamos começar por eles:  

**O que não importa?**  
Fatores como o grau planejado de estudo e o grau de estudo da mãe, a quantidade de bens e banheiros em casa e a escola ser pública ou privada não mostraram correlação expressiva com o consumo precosse de álcool.

**O que importa?**  
- A situação parental: Ter pais separados (morar com apenas um deles) mostra uma pequena correlação, entre as pessoas nessa situação o consumo de álcool é 9% maior que na base geral.
- Sentimento com o corpo e bullying: Parece existir uma correlação entre a satisfação com o corpo e a prática de bullying, 67% dos que disseram estar satisfeitos ou muito satisfeitos com o corpo praticam bullying - e 70% dos que fizeram bullying já ingeriram álcool. Ter sofrido bullying, contudo, não parece ter correlação.
- Companhia: A companhia importa e muito! Entre entrevistados que não possuem amigos que bebem, 78% desses também não bebem. Porém, quando todos os amigos bebem, esse número é de apenas 13%.

##Cigarro e outras drogas

Entre os que nunca consumiram álcool de maneira precoce, apenas 3% já experimentaram cigarro. Entre os que já consumiram álcool, esse número sobe para 32%. Contudo, os respondentes que não ingeriram álcool tendem a começar a fumar mais precocemente (dados e cruzamentos na Parte 3 do índice).

Essa correlação também existe com outras drogas, porém bem menor. Entre os que já ingeriram álcool, apenas 15% já usaram outras drogas. 

O mais interessante é que a correlação entre os menores de idade que já usaram cigarro e usaram outras drogas é a maior de todas, 37% dos respondentes. E aqueles que não ingeriram álcool também tendem a experimentar drogas mais cedo.

O cigarro precoce parece ser uma porta de entrada maior para outras drogas do que o álcool. Contudo, o álcool precoce também parece influenciar o menor de idade a fumar.
